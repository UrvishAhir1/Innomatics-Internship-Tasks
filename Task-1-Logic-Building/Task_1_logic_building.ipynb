## Task 1 – User Login Check

This task simulates a simple authentication system that verifies user credentials.

```python
# User Login Check

def check_login(input_usernmae, Input_password):
    username = "admin"
    password = "1234"

    if input_usernmae == username and Input_password == password:
        return "Login Successful"
    return "Invalid Credentials"

# User Input
uname = input("Enter user name: ")
upass = input("Enter user password: ")

result = check_login(uname, upass)

print(result)
```

## Task 2 – Pass / Fail Analyzer

This task analyzes student marks and counts the number of students who passed or failed.

```python
# Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Checking each student's marks
for score in marks:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Displaying final result
print("Total Students: ", len(marks))
print("Students Passed:", pass_count)
print("Students Failed:", fail_count)
```

## Task 3 – Simple Data Cleaner

This program cleans and standardizes a list of user names by:
- Removing extra spaces
- Converting all names to lowercase

```python
# Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Cleaning each name
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Displaying both original and cleaned list
print("Original Names: ", names)
print("Cleaned Names: ", cleaned_names)
```

## Task 4 – Message Length Analyzer

This program calculates the length of each message and flags messages
that are longer than 10 characters.

```python
# Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

# Analyzing each message
for message in messages:
    message_length = len(message)
    
    print(f"Message : {message}")
    print(f"Length  : {message_length}")
    
    # Checks the message lenght and printing message
    if message_length > 10:
        print("Status  : Long message detected")
    else:
        print("Status  : Short message")
    
    print("\n")
```

## Task 5 – Error Message Detector

This program scans system logs and counts how many times
an "ERROR" entry appears.

```python
# Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Checking each log entry
for entry in logs:
    if entry == "ERROR":
        error_count += 1

# Displaying result
print("Total ERROR detected:", error_count)
```
