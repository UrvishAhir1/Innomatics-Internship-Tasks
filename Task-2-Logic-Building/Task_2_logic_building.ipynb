 ## Problem 1 – Unique Words in a Sentence

This program identifies unique words from a given sentence and displays:
- The total number of unique words
- The unique words themselves

```python
sentence = "Python is easy and python is powerful" # We can also use input()

# Convert to lowercase for consistency
words = sentence.lower().split()

# Using set for keep only unique words 
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

## Problem 3 – Find Maximum and Minimum Values

This program finds the maximum and minimum values from a list of numbers
using basic loop comparison logic.

```python
numbers = [45, 22, 89, 10, 66]

# initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# loop through numbers
for num in numbers:
    if num > maximum: # check num is greater then max
        maximum = num
    if num < minimum:
        minimum = num # check num is smaller then mix

print("Numbers:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

## Problem 4 – Count Products Above a Price Threshold

This program counts how many products have a price greater than 1000.

```python
prices = [450, 1200, 899, 1500, 300]
count = 0  # stores number of products prices > 1000

# loop through each price
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

## Problem 5 – Calculate Attendance Percentage

This program calculates the attendance percentage
based on a list containing "P" (Present) and "A" (Absent).

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0  # stores number of present days

# count number of "P"
for present in attendance:
    if present == "P":
        present_count += 1

# calculate percentage
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

## Problem 7 – Count Character Frequency

This program counts how many times each character appears
in a given string.

```python
text = "Python files are stored as py"

# convert to lowercase
text = text.lower()

char_count = {}

# loop through each character
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

## Problem 10 – Calculate Average Marks

This program calculates the average marks of a student
from a list of marks.

```python
marks = [10, 20, 30, 40]

sum = 0   # initialize total

# calculate total using loop
for mark in marks:
    sum += mark

# calculate average
avg_marks = sum / len(marks)

print("Average Marks:", avg_marks)
```
