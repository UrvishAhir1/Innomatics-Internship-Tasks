## Problem 1: Mobile Recharge Validation System

**Problem Statement:** Create a function to validate a mobile recharge request.

**Rules:**
- Recharge amount must be ≥ ₹50
- Valid plans: 199, 299, 399, 599
- User should be allowed to retry if invalid

**Real-Time Use:**
- Telecom recharge apps
- Payment gateways

**Hint:**
- Use function for validation
- Use while loop for retry

```python
# Function to validate mobile recharge
def validate_recharge():
    valid_plans = [199, 299, 399, 599]  # Available recharge plans

    while True:  # Retry until valid input
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:  # Minimum amount check
            print("Recharge amount must be at least ₹50.")

        elif amount not in valid_plans:  # Plan validation
            print("Invalid plan. Available plans:", valid_plans)

        else:
            print("Recharge successful!")
            break

validate_recharge()
```

## Problem 2: Inventory Reorder Alert System

**Problem Statement:** Given a dictionary of products and stock quantities, check stock status.

**Requirements:**
- Print "Reorder Alert" if stock < 15
- Print "Stock OK" otherwise

**Real-Time Use:**
- Warehouse systems
- Retail management

**Hint:**
- Use function
- Use for loop on dictionary


```python
# Function to check inventory stock
def check_inventory(products):
    
    for product, stock in products.items():
        if stock < 15:  # Low stock condition
            print(product, "Reorder Alert")
        else:  # Sufficient stock
            print(product, "Stock OK")

inventory = {
    "Mobile": 10,
    "Laptop": 25,
    "Headphones": 12,
    "Charger": 18
}

check_inventory(inventory)
```

## Problem 3: Student Result Processing System

**Problem Statement:** Create a function that processes student marks and determines result.

**Requirements:**
- Accepts a list of student marks
- Calculates average marks
- Prints Pass if average ≥ 50 else Fail

**Real-Time Use:**
- Online exam systems
- Evaluation portals

**Hint:**
- Use loop to calculate total
- Return result from function

```python
# Function to process student result
def process_result(marks):
    
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)  # Calculate average
    
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result  # Return both values


student_marks = [60, 45, 70, 55]

avg, result = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", result)
```

## Problem 4: Cab Fare Estimator with Retry Option

**Problem Statement:** Create a cab fare calculator with peak hour pricing and retry option.

**Requirements:**
- Base fare = ₹50
- ₹12 per km
- Peak hour → 25% extra
- Allow retry using while loop

**Real-Time Use:**
- Ride-hailing apps
- Transport systems

**Hint:**
- Function for fare calculation
- While loop for retry

```python
# Function to calculate cab fare
def calculate_fare():
    base_fare = 50
    rate_per_km = 12

    while True:  # Retry loop
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        fare = base_fare + (distance * rate_per_km)  # Normal fare

        if peak == "yes":  # Add 25% in peak hour
            fare += fare * 0.25

        print(f"Estimated Fare: ₹{fare}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break

calculate_fare()
```

## Problem 5: Employee Attendance Eligibility Checker

**Problem Statement:** Create a function that checks employee attendance eligibility.

**Requirements:**
- Accepts attendance list ("P", "A")
- Calculates attendance percentage
- Returns Eligible if ≥ 75% else Not Eligible

**Real-Time Use:**
- HR payroll systems
- Internship tracking

**Hint:**
- Use loop to count present days
- Use return statement

```python
# Function to check attendance eligibility
def check_attendance(attendance):
    present_count = 0
    
    # Count present days
    for day in attendance:
        if day == "P":
            present_count += 1
    
    percentage = (present_count / len(attendance)) * 100  # Calculate percentage
    
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    return percentage, status

attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

percentage, result = check_attendance(attendance_list)

print("Attendance Percentage:", percentage, "%")
print("Attendance Result:", result)
```

## Problem 6: Password Strength Checker

**Problem Statement:** Create a function to check password strength.

**Rules:**
- Minimum length = 8
- Must contain at least one digit
- Must contain at least one special character (@#$)

**Real-Time Use:**
- Login systems
- Security validation

**Hint:**
- Use loop over string
- Use conditions inside function

```python
# Function to check password strength
def check_password(password):
    
    has_digit = False
    has_special = False
    
    # Check length
    if len(password) < 8:
        return "Minimum 8 characters required"
    
    # Loop through each character
    for ch in password:
        
        if ch.isdigit():  # Check for digit
            has_digit = True
        
        if ch in "@#$":  # Check for special character
            has_special = True
    
    # Final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"

user_password = input("Enter password: ")
result = check_password(user_password)

print("Password:", user_password)
print("Password Status:", result)
```
